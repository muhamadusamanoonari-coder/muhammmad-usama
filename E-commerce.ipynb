```python
import csv
import os


products = {}


order = []




def setup_product_file():
    """If 'products.csv' doesn't exist, this creates a simple one."""
    if not os.path.exists('products.csv'):
        print("products.csv not found. Creating a sample file for you.")
        with open('products.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['id', 'name', 'price'])
            writer.writerow([1, 'Laptop', 1200])
            writer.writerow([2, 'Phone', 800])
            writer.writerow([3, 'Headphones', 150])

def load_products():
    """Reads the products from the CSV file into our 'products' dictionary."""
    try:
        with open('products.csv', mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                product_id = int(row['id'])
                products[product_id] = {
                    'name': row['name'],
                    'price': float(row['price'])
                }
    except FileNotFoundError:
        print("ERROR: Cannot find products.csv!")
    except Exception:
        print("ERROR: The CSV file might be formatted incorrectly.")

def show_products():
    """Prints a list of all products for the user to see."""
    print("\n--- Available Products ---")
    print(f"{'ID':<5} {'Name':<20} {'Price'}")
    print("-" * 35)
    for product_id, details in products.items():
        print(f"{product_id:<5} {details['name']:<20} ${details['price']:.2f}")
    print("-" * 35)

def add_to_order():
    """Asks the user what they want to buy and adds it to the order."""
    try:
        product_id = int(input("Enter product ID to add: "))
        quantity = int(input("Enter quantity: "))

        if product_id not in products:
            print("Sorry, that product ID does not exist.")
        elif quantity <= 0:
            print("Quantity must be a positive number.")
        else:
            product = products[product_id]
            order_item = {
                'name': product['name'],
                'quantity': quantity,
                'price': product['price']
            }
            order.append(order_item)
            print(f"Added {quantity} x {product['name']} to your order.")
    except ValueError:
        print("Invalid input. Please enter numbers for ID and quantity.")

def show_order():
    """Displays the items in the order and calculates the total cost."""
    print("\n--- Your Order ---")
    if not order:
        print("Your order is empty.")
        return

    total_cost = 0.0
    for item in order:
        subtotal = item['price'] * item['quantity']
        print(f"- {item['name']} (x{item['quantity']}): ${subtotal:.2f}")
        total_cost += subtotal

    print("--------------------")
    print(f"Total: ${total_cost:.2f}")



setup_product_file()
load_products()

print("\nWelcome to the Simple Shop!")


while True:
    print("\n--- Menu ---")
    print("1. View products")
    print("2. Add item to order")
    print("3. View your order")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        show_products()
    elif choice == '2':
        add_to_order()
    elif choice == '3':
        show_order()
    elif choice == '4':
        print("Goodbye!")
        break  # This stops the loop.
    else:
        print("Invalid choice. Please pick a number from 1 to 4.")
```
